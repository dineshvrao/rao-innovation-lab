```python
# Databricks notebook source

spark.sql("""
CREATE SCHEMA IF NOT EXISTS demo_db
""")

spark.sql("""
CREATE TABLE IF NOT EXISTS demo_db.customers (
  customer_id INT,
  customer_name STRING,
  region STRING,
  created_date DATE
)
USING DELTA
""")

spark.sql("""
CREATE TABLE IF NOT EXISTS demo_db.orders (
  order_id INT,
  customer_id INT,
  product_name STRING,
  quantity INT,
  unit_price DOUBLE,
  order_date DATE
)
USING DELTA
""")

spark.sql("""
INSERT INTO demo_db.customers VALUES
(1, 'Alice', 'East', current_date()),
(2, 'Bob', 'West', current_date()),
(3, 'Charlie', 'South', current_date())
""")

spark.sql("""
INSERT INTO demo_db.orders VALUES
(101, 1, 'Laptop', 1, 1200.00, current_date()),
(102, 2, 'Monitor', 2, 300.00, current_date()),
(103, 3, 'Keyboard', 3, 80.00, current_date())
""")

display(spark.sql("SELECT * FROM demo_db.customers"))
display(spark.sql("SELECT * FROM demo_db.orders"))
```
